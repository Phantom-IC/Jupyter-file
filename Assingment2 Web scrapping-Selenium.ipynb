```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd 
import time
```

```python
jb_title=[]
jb_exp=[]
compy_nm=[]
loca=[]
url='https://www.naukri.com/'
driver=webdriver.Chrome(r'C://chromedriver.exe')
driver.get(url)
sc=driver.find_element_by_name('keyword')
sc.send_keys('data analyst')
sc1=driver.find_element_by_name('location')
sc1.send_keys('bangalore')
srch=driver.find_element_by_xpath("//div[@class='search-btn']/button")
srch.click()
try:
    title = WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.XPATH, "//a[@class='title fw500 ellipsis']"))
    )
    exp = WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.XPATH, "//li[@class='fleft grey-text br2 placeHolderLi experience']"))
    )
    comny_nme = WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.XPATH, "//a[@class='subTitle ellipsis fleft']"))
    )
    location = WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.XPATH, "//li[@class='fleft grey-text br2 placeHolderLi location']"))
    )
    for i in title[1:11]:
        jb_title.append(i.text)
    for i in exp[1:11]:
        jb_exp.append(i.text)
    for i in comny_nme[1:11]:
        compy_nm.append(i.text)
    for i in location[1:11]:
        loca.append(i.text)
finally:
    driver.quit()
time.sleep(8)
driver.quit()

df=pd.DataFrame({'Job Title':jb_title,'Job Location':loca,
              'Company name':compy_nm,'Experience Required':jb_exp,})
print(df)
df.to_csv('Naukri-job',index=False)
```

```python
jb_title=[]
jb_exp=[]
compy_nm=[]
loca=[]
url='https://www.naukri.com/'
driver=webdriver.Chrome(r'C://chromedriver.exe')
driver.get(url)
sc=driver.find_element_by_name('keyword')
sc.send_keys('data Scientist')
srch=driver.find_element_by_xpath("//div[@class='search-btn']/button")
srch.click()
try:
    path = WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.XPATH, "//span[@class='ellipsis fleft']"))
    )
    for i in path:
        if i.text=='Delhi / NCR':
            i.click()
            break
    title = WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.XPATH, "//a[@class='title fw500 ellipsis']"))
    )
    
    location = WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.XPATH, "//li[@class='fleft grey-text br2 placeHolderLi location']"))
    )
    cmpy_name = WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.XPATH, "//a[@class='subTitle ellipsis fleft']"))
    )
    exp = WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.XPATH, "//li[@class='fleft grey-text br2 placeHolderLi experience']"))
    )
    for i in title[0:11]:
        jb_title.append(i.text)
    for i in exp[0:11]:
        jb_exp.append(i.text)
    for i in comny_nme[0:11]:
        compy_nm.append(i.text)
    for i in location[0:11]:
        loca.append(i.text)
except:
    h=[]
    v=[]
    location2=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']")
    cpn_nm=driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']")
    for i in location2[0:11]:
        h.append(i.text)
    for i in cpn_nm[0:11]:
        v.append(i.text)
    driver.quit()
time.sleep(5)
driver.quit()
```

```python
df=pd.DataFrame({})
df['Job title']=jb_title
df['Job location']=h
df['Company name']=v
df['Required Experience']=jb_exp
df
df.to_csv('Data Scientist', index=False)
```

```python
url='https://www.glassdoor.com/Job/noida-data-scientist-jobs-SRCH_IL.0,5_IC4477468_KO6,20.htm'
driver=webdriver.Chrome(r'C://chromedriver.exe')
driver.get(url)
name=[]
ratings=[]
age=[]
try:
    cmpy_name = WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.XPATH, "//div[@class='d-flex justify-content-between align-items-start']"))
    )
    rating= WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.XPATH, "//span[@class='css-19pjha7 e1cjmv6j1']"))
    )
    days_ago= WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.XPATH, "//div[@class='d-flex align-items-end pl-std css-mi55ob']"))
    )
    exp = WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.XPATH, "//li[@class='fleft grey-text br2 placeHolderLi experience']"))
    )
    for i in cmpy_name[0:11]:
        name.append(i.text)
    for i in rating[0:11]:
        ratings.append(i.text)
    for i in days_ago[0:11]:
        age.append(i.text)
except:
    n=[]
    r=[]
    a=[]
    names=driver.find_elements_by_xpath("//div[@class='d-flex justify-content-between align-items-start']")
    rating2=driver.find_elements_by_xpath("//span[@class='css-19pjha7 e1cjmv6j1']")
    days_ago2=driver.find_elements_by_xpath("//div[@class='d-flex align-items-end pl-std css-mi55ob']")
    for i in days_ago2[0:11]:
        a.append(i.text)
    for i in names[0:11]:
        n.append(i.text)
    for i in rating2[0:11]:
        r.append(i.text)
time.sleep(6)
driver.quit()
```

```python
brands=[]
descriptions=[]
prices=[]
discounts=[]
url= 'https://www.flipkart.com/'
driver=webdriver.Chrome(r'C://chromedriver.exe')
driver.get(url)
sc=driver.find_element_by_name('q')
sc.send_keys('sunglasses')
sc1=driver.find_element_by_xpath("//button[@class='L0Z3Pu']")
sc1.click()
try:
    brand = WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.XPATH, "//div[@class='_2WkVRV']"))
    )
    description= WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.XPATH, "//a[@class='IRpwTa']"))
    )
    price= WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.XPATH, "//div[@class='_30jeq3']"))
    )
    discount= WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.XPATH, "//div[@class='_3Ay6Sb']"))
    )
    url=driver.find_element_by_xpath("//nav[@class='yFHi8N']").get_attribiute('href')
    url.click()
    for i in brand[0:37]:
        brands.append(i.text)
    for i in description[0:37]:
        descriptions.append(i.text)
    for i in price[0:37]:
        prices.append(i.text)
    for i in discount[0:37]:
        discounts.append(i.text)
except:
    a_b=[]
    b_b=[]
    c_b=[]
    d_b=[]
    for i in range(0,6):
        for j in driver.find_elements_by_xpath("//div[@class='_2WkVRV']"):
            a_b.append(j.text)
        for k in driver.find_elements_by_xpath("//a[@class='IRpwTa']"):
            b_b.append(k.text)
        for l in driver.find_elements_by_xpath("//div[@class='_30jeq3']"):
            c_b.append(l.text)
        for m in driver.find_elements_by_xpath("//div[@class='_3Ay6Sb']"):
            d_b.append(m.text)
        time.sleep(5)
        url=driver.find_element_by_xpath("//nav[@class='yFHi8N']/a").get_attribute("href")
        driver.get(url)
time.sleep(7)
driver.quit()
```

```python
df=pd.DataFrame({})
df['Brands']=a_b[0:100]
df['Descriptions']=b_b[0:100]
df['Prices']=c_b[0:100]
df['Discounts']=d_b[0:100]
df.to_csv('Fetched')
df
```
